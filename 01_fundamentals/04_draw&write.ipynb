# Draw&Write

OpenCV has a number of drawing functions you can use to draw various shapes, including polygons of irregular shapes, but the three most common OpenCV drawing functions you will see are:

- cv2.line: Draws a line on image, starting at a specified (x, y)-coordinate and ending at another (x, y)-coordinate
- cv2.circle: Draws a circle on an image specified by the center (x, y)-coordinate and a supplied radius
- cv2.rectangle: Draws a rectangle on an image specified by the top-left corner and bottom-right corner (x, y)-coordinates

# Line

```python
# import the necessary packages
import numpy as np
import cv2
# initialize our canvas as a 300x300 pixel image with 3 channels
# (Red, Green, and Blue) with a black background
Blank = np.zeros((600, 300, 3), dtype="uint8")
cv2.imshow('Blank',Blank)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# draw a green line from the top-left corner of our canvas to the bottom-right
green = (0, 255, 0)
topleft=cv2.line(Blank, (0, 0), (300, 600), green)
cv2.imshow("top-left", topleft)
cv2.waitKey(0)
# draw a 5 pixel thick red line from the top-right corner to the bottom-left
red = (0, 0, 255)
topright=cv2.line(Blank, (0, 600), (300, 0), red, 5)
cv2.imshow("topright", topright)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
