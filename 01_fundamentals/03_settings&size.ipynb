# Getting and Setting Pixels

```python
print(img.shape)
```

In OpenCV, pixels are accessed by their (x, y)-coordinates. The origin, (0, 0), is located at the top-left of the image. OpenCV images are zero-indexed, where the x-values go left-to-right (column number) and y-values go top-to-bottom (row number)

To access a particular image pixel, all we need to do is pass in the (x, y)-coordinates as image[y, x]

```python
(h, w) = img.shape[:2]
print(h,w)
```

```python
img_mammotty= img[0:664, 50:550]
cv2.imshow('Mamotty', img_mammotty)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img_mohanlal= img[0:664, 600:1053]
cv2.imshow('Mohanlal', img_mohanlal)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# write an Image

```python
cv2.imwrite("Mammotty_cropped.jpg", img_mammotty)
cv2.imwrite("Mohanlal_cropped.jpg", img_mohanlal)
```
